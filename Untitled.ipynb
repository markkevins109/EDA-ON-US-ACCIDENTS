```python
missing_percentages=df.isnull().sum().sort_values(ascending=False)/len(df)*100
```

```python
missing_percentages
```

```python
missing_percentages=missing_percentages[missing_percentages>0]
missing_percentages
```

```python
missing_percentages.plot(kind="barh")
```

```python
columns_to_drop = missing_percentages[missing_percentages > 50]
df = df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)/len(df)*100
```

```python
cities=df.City.unique()
len(cities)
```

```python
cities_by_accident=df.City.value_counts()
cities_by_accident
```

```python
cities_by_accident[:20].plot(kind='bar')
```

```python
high_accident_cities=cities_by_accident[cities_by_accident>=1000]
low_accident_cities=cities_by_accident[cities_by_accident<1000]
print("high accidents count",len(high_accident_cities))
print("high accident percentage",len(high_accident_cities)/len(cities)*100)
len(low_accident_cities)/len(cities)*100
```

```python
import seaborn as sns
sns.histplot(high_accident_cities,log_scale=True)
```

```python
sns.histplot(low_accident_cities,log_scale=True)
```

```python
low_accident_cities[low_accident_cities==1]
```

```python
high_s = list(high_accident_cities.keys())
state_list = df['State'].tolist()
state_unique_list = list(set(state_list))
state_unique_list_counts = [state_list.count(s) for s in state_unique_list]

import matplotlib.pyplot as plt
explode = [0.1 if state == high_s[0] else 0 for state in state_unique_list]

plt.figure(figsize=(10, 7))
plt.pie(state_unique_list_counts, labels=state_unique_list, autopct='%0.01f%%', explode=explode)
plt.title('Accidents Distribution by State')
plt.show()
```
